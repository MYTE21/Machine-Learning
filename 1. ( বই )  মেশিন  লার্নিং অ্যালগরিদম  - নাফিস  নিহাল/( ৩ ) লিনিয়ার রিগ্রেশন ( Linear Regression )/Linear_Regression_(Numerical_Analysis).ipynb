```
import pandas as pd
result = pd.read_csv('LinearRegression_NumericalAnalysis.csv')
print(result)
```

```
import matplotlib.pyplot as plt
plt.scatter(result.X, result.Y)
plt.show()
```

```
import numpy as np
X_avg = np.mean(result.X)
Y_avg = np.mean(result.Y)
XY_avg = np.mean(result.X * result.Y)
X_avg_sq = X_avg**2
X_sq_avg = np.mean(result.X**2)
print(X_avg)
print(Y_avg)
print(XY_avg)
print(X_avg_sq)
print(X_sq_avg)
```

```
m = (((X_avg*Y_avg)-(XY_avg))/(X_avg_sq-X_sq_avg))
print(m)
```

```
c = Y_avg - m*X_avg
print(c)
y = m*result.X + c
```

```
plt.plot(result.X, y, '-r', label='y = m*result.X - c')
plt.show()
```

```
down = np.sum((result.Y - Y_avg)**2)
Y_pred = y
up = np.sum((Y_pred - Y_avg)**2)
R_sq = up/down
print(R_sq)
```

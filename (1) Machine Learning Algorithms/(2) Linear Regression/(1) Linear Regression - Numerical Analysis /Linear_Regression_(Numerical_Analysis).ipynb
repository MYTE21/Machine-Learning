**Modify data**

```
data = data.rename(
              columns=
              {
                'Size': 'X',
                'Price': 'Y'
              })

data.insert(2, 'XY', X*Y)
data.insert(3, 'X^2', X**2)

print(data)
```

**Equations**


*   **Best Fit Line** : y = mx + c
*   <img src="https://latex.codecogs.com/svg.latex?\Large&space;m=\frac{\bar{x} \cdot \bar{y} - \overline{xy}}{(\bar{x})^2 - \overline{x^2}}"/>
*  <img src="https://latex.codecogs.com/svg.latex?\Large&space;c=\bar{y}-m\bar{x}"/>

```
import numpy as np
X_avg = np.mean(data.X)
Y_avg = np.mean(data.Y)
XY_avg = np.mean(data.XY)
X_avg_sq = X_avg**2
X_sq_avg = np.mean(data.X**2)
print('X_avg : ', X_avg)
print('Y_avg : ', Y_avg)
print('XY_avg : ', XY_avg)
print('X_avg_sq : ', X_avg_sq)
print('X_sq_avg : ', X_sq_avg)
```

**Find 'm'**

```
m = (((X_avg*Y_avg)-(XY_avg))/(X_avg_sq-X_sq_avg))
print('m : ', m)
```

**Find 'c'**

```
c = Y_avg - m*X_avg
print('c : ', c)
```

**Find 'y' for all 'x'**

```
y = m*data.X + c
print('y : \n\n', y)
```

**Plot result**

```
plt.plot(data.X, y, label='y = m*data.X - c')
plt.title('Pizza Size vs Price')
plt.xlabel('Size in inches')
plt.ylabel('Price in taka')
plt.show()
```

**Accuracy measure :**


*   R_sq = R-Squared Value
*   <img src="https://latex.codecogs.com/svg.latex?\Large&space;R^2=\frac{\sum_{i=1}^{n}(\hat{y_i} - \bar{y})^2}{\sum_{i=1}^{n}({y_i} - \bar{y})^2}"/>



```
Y_pred = y
up = np.sum((Y_pred - Y_avg)**2)
down = np.sum((data.Y - Y_avg)**2)
R_sq = up/down
print('R-Squared Value : ', R_sq)
```
